```python
import multiprocessing as mp
import tskit

def get_newick_safe(tree_idx):
    global sub_ts  # global must be set outside pool
    tree = sub_ts.at(tree_idx)
    if tree.num_roots == 1:
        return tree.newick()
    return None

# Only on Unix:
mp.set_start_method("fork", force=True)

# sub_ts = tskit.load("./sample.trees")
tree_indexes = list(range(sub_ts.num_trees))

with mp.Pool(mp.cpu_count()) as pool:
    results = pool.map(get_newick_safe, tree_indexes)

# Filter out None values
nwk_list = [r for r in results if r is not None]
```
